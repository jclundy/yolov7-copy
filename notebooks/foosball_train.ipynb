```python
# includes
import os
from sklearn.model_selection import train_test_split
import shutil
```

```python
## Split the data
```

```python
training_image_folder = "/home/joseph/Documents/project/videos/foosball_videos/20221004_220424_jpg"
training_annotations_folder = "/home/joseph/Documents/project/videos/foosball_videos/20221004_220424_labels"

# Read images and annotations
images = [os.path.join(training_image_folder, x) for x in os.listdir(training_image_folder)]
annotations = [os.path.join(training_annotations_folder, x) for x in os.listdir(training_annotations_folder) if x[-3:] == "txt"]

images.sort()
annotations.sort()

# Split the dataset into train-valid-test splits 
train_images, val_images, train_annotations, val_annotations = train_test_split(images, annotations, test_size = 0.2, random_state = 1)
val_images, test_images, val_annotations, test_annotations = train_test_split(val_images, val_annotations, test_size = 0.5, random_state = 1)
```

```python
print(len(annotations))
print(len(images))
```

```python
data_set_root = "../Foosball_Dataset"
image_folder = data_set_root + "/images"
label_folder = data_set_root + "/labels"

image_train_folder = image_folder + "/train"
image_test_folder = image_folder + "/test"
image_val_folder = image_folder + "/val"

label_train_folder = label_folder + "/train"
label_test_folder = label_folder + "/test"
label_val_folder = label_folder + "/val"

folder_list = [image_train_folder, image_test_folder, image_val_folder, label_train_folder, label_test_folder, label_val_folder]
for folder in folder_list:
    if not os.path.exists(folder):
        os.makedirs(folder)
```

```python
#Utility function to move images 
def move_files_to_folder(list_of_files, destination_folder):
    for f in list_of_files:
        try:
            shutil.move(f, destination_folder)
        except:
            print(f)
            assert False

```

```python
# Move the splits into their folders
move_files_to_folder(train_images, image_train_folder)
move_files_to_folder(val_images, image_val_folder)
move_files_to_folder(test_images, image_test_folder)
move_files_to_folder(train_annotations, label_train_folder)
move_files_to_folder(val_annotations, label_test_folder)
move_files_to_folder(test_annotations, label_val_folder)
```
